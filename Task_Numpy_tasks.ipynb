#### 56. Generate a generic 2D Gaussian-like array (★★☆) 
(**hint**: np.meshgrid, np.exp)

````python
import numpy as np
A,B=np.meshgrid(np.linspace(-1,1,10),np.linspace(-1,1,10))
D=np.sqrt(A*A+B*B)
sigma,mu=1.0, 0.0
G=np.exp(-( (D-mu)**2 / ( 2.0 * sigma**2 ) ) )
print(G)
````
